```python
#Joins
employee_data = {'employee_id': [1,2,3,4,5],
              'name':['Babatunde', 'Ozy', 'Muaz','Oladehinde','Moses']}
df = pd.DataFrame(employee_data, columns = ['employee_id', 'name'])
```

```python
df
```

```python
sales_data = {'employee_id': [3,4,5,6],
             'total_sales': [2345, 1234, 4532, 2890]}
sales_dataframe = pd.DataFrame(sales_data, columns=['employee_id', 'total_sales'])
```

```python
sales_dataframe
```

```python
#innerjoin
pd.merge(df, sales_dataframe, on='employee_id')
```

```python
#outer join
pd.merge(df, sales_dataframe, on='employee_id', how='outer')
```

```python
#left joins
pd.merge(df, sales_dataframe, on='employee_id', how='left')
```

```python
#right joins
pd.merge(df, sales_dataframe, on='employee_id', how='right')
```
